```python
def maxpair(n, a):
    max_index1 = 0
    max_index2 = 0
    ans = 0
    for i in range(len(a)):
        if(a[i]>a[max_index1]):
            max_index1=i
    for j in range(len(a)):
        if(a[j]>a[max_index2] and j!=max_index1):
            max_index2=j
    ans = a[max_index1]*a[max_index2]
    return ans




n = float(input())
a = list(map(float, input().split()))

result = maxpair(n, a)
print(result)
```

```python
def app(arr,total,n):
    if(n==0 or total==0):
        print('1 ',total)
        return total
    if(arr[n-1]<=total):
        print('2 ',total)
        return max(app(arr,total-arr[n-1],n-1),app(arr,total,n-1))

    elif(arr[n-1]>total):
        print('3 ',total)
        return app(arr,total,n-1)



n=int(input())
arr=list(map(int,input().split()))
arr.sort()
sum1=sum(arr)
total=sum1//2
to=sum1-total
print(to-(total-app(arr,total,n)))
```
