```python
n,k=list(map(int,input().split()))
if(k>n):
    print('NO')
elif(k==0):
    print('YES')
else:
    li=[0]*(n+1)
    li[0]=li[1]=10**4
    prime=[]
    flag=0
    for i in range(2,(n//2)+1):
        #print(i)
        if(li[i]==0):
            #print(i)
            for j in range(i*2,n+1,i):
                #print(j)
                li[j]=1


    for i in range(len(li)):
        if(li[i]==0):
            prime.append(i)
    value=[]
    count=0
    for i in range(len(prime)-1):
        va=prime[i]+prime[i+1]+1
        if(va in prime):
            #print(va)
            count=count+1
        if(count>=k):
            flag=1
            break
    if(flag==1):
        print('YES')
    else:
        print('NO')
```

# ETF - Euler Totient Function

```python
#from math import gcd
for _ in range(int(input())):
    n=int(input())
    count=1
    def gcd(a,b):
        if (a == 0): 
            return b 
        return gcd(b % a, a) 
    
    for i in range(2,n):
        if(gcd(i,n)==1):
            count+=1
    print(count)
```

```python
def phi(n): 
    result = n;  
    p = 2;  
    while(p * p <= n): 
        if (n % p == 0):  
            while (n % p == 0): 
                n = int(n / p); 
            result -= int(result / p); 
        p += 1; 
    if (n > 1): 
        result -= int(result / n); 
    return result; 
  
for _ in range(int(input())):
    n=int(input())
    print(phi(n))
      
```

# LCMSUM - LCM Sum

```python
from math import gcd
for _ in range(int(input())):
    n=int(input())
    count=0
    for i in range(1,n+1):
        if(n%i==0):
            count=count+n
        else:
            count=count+(i*n)//gcd(i,n)
    print(count)
            
        
```
