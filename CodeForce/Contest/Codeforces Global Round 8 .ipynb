# A. C+= 

```python
for _ in range(int(input())):
    a,b,n=list(map(int,input().split()))
    count=0
    flag=0
    if(a>b):
        flag=1
    else:
        flag=0
    while(a<=n and b<=n):
        if(flag==0):
            a+=b
            count+=1
            flag=1
        elif(flag==1):
            b+=a
            count+=1
            flag=0
            
    print(count)
            
        
```

```python
n=int(input())
if(n==2):
    print(10)
    print(0,1)
    print(0,2)
    for i in range(1,3):
        print(i,0)
        print(i,1)
        print(i,2)
        print(i,3)
if(n==1):
    print(8)
    print(0,1)
    print(0,2)
    for i in range(1,3):
        print(i,0)
        print(i,1)
        print(i,2)

if(n==3):
    print(12)
    print(0,1)
    print(0,2)
    for i in range(1,3):
        print(i,0)
        print(i,1)
        print(i,2)
        print(i,3)
    print(3,2)
    print(3,3)
else:
    flag=0
    if(n%2==0):
        value=n+n+4
        flag=1
    else:
        x=n+1
        value=x+(x-2)+4+1
    print(value)
    if(flag==1):
        print(0,n-3)
        print(0,n-2)
        for i in range(1,(n//2)+1):
            print(i,n-4)
            print(i,n-3)
            print(i,n-2)
            print(i,n-1)
        print((n//2)+1,n-3)
        print((n//2)+1,n-2)
    else:
        print(0,n-3)
        print(0,n-2)
        for i in range(1,((n-1)//2)+1):
            print(i,n-4)
            print(i,n-3)
            print(i,n-2)
            print(i,n-1)
        print(((n-1)//2)+1,n-3)
        print(((n-1)//2)+1,n-2)
        print(((n-1)//2)+1,n-1)
        print(((n-1)//2)+2,n-2)
        print(((n-1)//2)+2,n-1)
```
